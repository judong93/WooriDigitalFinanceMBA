 #### ※ EXTRA EXPLANATION ON COPY ISSUE

 ##### shallow copy & deep copy in Python

```python
l1 = [1,2,3,4]
l2 = l1 # alias
l2.append(5)
print(l1) # l1 changed
print(l2)
```

```python
import copy

l1 = [1,2,3,4]
l2 = copy.copy(l1) # (shallow) copy
l2.append(5)
print(l1) # l1 not changed
print(l2)
```

```python
l1 = [[1,2], [2,3], [3,4]]
l2 = copy.copy(l1) # shallow copy
l2[0][0] = 0
print(l1) # l1 changed
print(l2)
```

```python
l1 = [[1,2], [2,3], [3,4]]
l2 = copy.deepcopy(l1) # deep copy
l2[0][0] = 0
print(l1) # l1 not changed
print(l2)
```

 ##### shallow copy & deep copy in Pandas

```python
x = {'name': ['kim', 'park'],
     'info': [[160, 59], [170, 70]]}
df1 = pd.DataFrame(x)
df2 = df1.copy(deep=False) # with the option 'deep=False', make alias of the DataFrame object
# copy_s = df_x # same as above
df2.loc[0, 'name'] = 'lee'
print(df1.loc[0, 'name']) # df1 changed
print(df2.loc[0, 'name'])
```

```python
x = {'name': ['kim', 'park'],
     'info': [[160, 59], [170, 70]]}
df1 = pd.DataFrame(x)
df2 = df1.copy(deep=True) # with the option 'deep=True', shallow-copy the DataFrame object
df2.loc[0, 'name'] = 'lee'
print(df1.loc[0, 'name']) # df1 not changed
print(df2.loc[0, 'name'])
```

```python
x = {'name': ['kim', 'park'],
     'info': [[160, 59], [170, 70]]}
df1 = pd.DataFrame(x)
df2 = df1.copy(deep=True) # with the option 'deep=True', shallow-copy the DataFrame object
df2.loc[0, 'info'][0] = 'new value'
print(df1.loc[0, 'info']) # df1 changed
print(df2.loc[0, 'info'])
```

```python
x = {'name': ['kim', 'park'],
     'info': [[160, 59], [170, 70]]}
df1 = pd.DataFrame(x)
df2 = pd.DataFrame(copy.deepcopy(df1.to_dict())) # to perfectly deep-copy the DataFrame object, create a new DataFrame object with the values extracted
df2.loc[0, 'info'][0] = 'new value'
print(df1.loc[0, 'info']) # df1 not changed
print(df2.loc[0, 'info'])
```

```python
def func(cell):
    list_rating = [pd.NA, pd.NA, pd.NA]
    for rating in cell:
        if rating.startswith('호텔스컴바인'):
            list_rating[0] = float(rating[6:])
        elif rating.startswith('트립어드바이저'):
            list_rating[1] = float(rating[7:-2]) * 2
        else:
            list_rating[2] = float(rating[4:-3])
    return list_rating

df_hotel['rating'] = df_hotel['rating'].apply(func)
df_hotel
```
